##Importing the libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## Data Collection and Pre-Processing

```python
movies_data = pd.read_csv('/content/movies-2.csv')
```

```python
#Printing the first 5 rows of the dataframe
movies_data.head()
```

```python
 #number of rows and colums in dataset
 movies_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
print(selected_features)
```

```python
#replacing the missing values with null string
for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
#combining all the five selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#converting the text data to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors=vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

## Cosine Similarity

```python
#getting the similarity scores using cosine similarity
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user
movie_name=input('Enter your favourite movie name: ')
```

```python
#creating the list with all movie names given in the data set
list_of_all_title=movies_data['title'].tolist()
print(list_of_all_title)
```

```python
#finding the close match for the movie name given by the user
find_close_match=difflib.get_close_matches(movie_name,list_of_all_title)
```

```python
print(find_close_match)
```

```python
#we will take only first value
close_match=find_close_match[0]
print(close_match)
```

```python
#findind the index of movie with title
index_of_movie=movies_data[movies_data.title==close_match]['index'].values[0]
```

```python
print(index_of_movie)
```

```python
#getting the list of similar movies
similarity_score=list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
#sorting the movies based on their similarity score
sorted_similar_movies=sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

```python
#printing the name of similar movies
print('Movies Suggested for you: \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=movies_data[movies_data.index==index]['title'].values[0]
  if(i<=20):
    print(i,'.', title_from_index)
    i+=1
```
